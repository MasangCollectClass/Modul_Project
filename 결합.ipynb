```python
import tensorflow as tf
import pickle
from tensorflow.keras.preprocessing.sequence import pad_sequences

# ğŸ“ ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ê²½ë¡œ
model_paths = {
    "ei": ("ei_bilstm_model.h5", "ei_tokenizer.pkl"),
    "ns": ("ns_bilstm_model.h5", "ns_tokenizer.pkl"),
    "tf": ("tf_bilstm_model.h5", "tf_tokenizer.pkl"),
    "jp": ("jp_bilstm_model.h5", "jp_tokenizer.pkl"),
}

# ğŸ“¦ ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¡œë”©
models = {}
tokenizers = {}
max_len = 300

for trait, (model_file, tokenizer_file) in model_paths.items():
    models[trait] = tf.keras.models.load_model(model_file)
    with open(tokenizer_file, "rb") as f:
        tokenizers[trait] = pickle.load(f)

# ğŸ”¤ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ í•¨ìˆ˜
def preprocess(text, tokenizer):
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')
    return pad

# ğŸ§  MBTI ì˜ˆì¸¡ í•¨ìˆ˜
def predict_mbti(text: str) -> str:
    result = ""
    label_pairs = {
        "ei": ("E", "I"),
        "ns": ("N", "S"),
        "tf": ("T", "F"),
        "jp": ("J", "P"),
    }
    for trait in ["ei", "ns", "tf", "jp"]:
        x = preprocess(text, tokenizers[trait])
        pred = models[trait].predict(x)[0][0]
        pos, neg = label_pairs[trait]
        label = pos if round(pred) == 1 else neg
        result += label
    return result


# ğŸ—ƒï¸ ì‚¬ìš©ì ì…ë ¥ ëˆ„ì  ë¦¬ìŠ¤íŠ¸
session_inputs = []

# â• ë¬¸ì¥ ëˆ„ì  í•¨ìˆ˜
def add_user_text(text: str):
    session_inputs.append(text)
    print(f"í˜„ì¬ ëˆ„ì  ë¬¸ì¥ ìˆ˜: {len(session_inputs)}")
    
    if len(session_inputs) >= 5:
        combined = ' '.join(session_inputs)
        mbti = predict_mbti(combined)
        print(f"[ì˜ˆì¸¡ëœ MBTI] â†’ {mbti}")
        return mbti
    else:
        print("ë¬¸ì¥ì„ 5ê°œ ì´ìƒ ì…ë ¥í•˜ë©´ MBTIë¥¼ ì˜ˆì¸¡í•©ë‹ˆë‹¤.")
        return None

# ğŸ§ª ì‚¬ìš© ì˜ˆì‹œ
add_user_text("í˜¼ì ìˆëŠ” ê²Œ í¸í•´ìš”.")
add_user_text("ì‚¬ëŒ ë§ì€ ê³³ì€ ê¸ˆë°© ì§€ì³ìš”.")
add_user_text("ê°ì • í‘œí˜„ì€ ì˜ ì•ˆ í•˜ëŠ” í¸ì´ì—ìš”.")
add_user_text("ê³„íšë³´ë‹¤ ì¦‰í¥ì„ ì¢‹ì•„í•´ìš”.")
add_user_text("ìƒê°ì´ ë§ì•„ì§€ëŠ” ë°¤ì´ ì‹«ì§€ ì•Šì•„ìš”.")  # â†’ ì—¬ê¸°ì„œ ì˜ˆì¸¡ë¨
```
